### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
##Write your code here

df_filtered = df.dropna(subset=['JobSatPoints_6', 'JobSatPoints_7', 'Employment'])
employment_counts = df_filtered['Employment'].value_counts()
num_categories = 10
employment_counts_top = employment_counts.head(num_categories - 1)
employment_counts_top['Other'] = employment_counts[num_categories - 1:].sum()

employment_mapping = {status: status if status in employment_counts_top.index else 'Other' for status in df['Employment'].unique()}
df_filtered['EmploymentGroup'] = df_filtered['Employment'].map(employment_mapping)
job_satisfaction_median = df_filtered.groupby('EmploymentGroup')[['JobSatPoints_6', 'JobSatPoints_7']].median()
fig, ax = plt.subplots(figsize=(8, 8))
bar_width = 0.5
employment_groups = np.arange(len(job_satisfaction_median))

bar1 = ax.bar(employment_groups, job_satisfaction_median['JobSatPoints_6'], bar_width, label='JobSatPoints_6', color='red')
bar2 = ax.bar(employment_groups, job_satisfaction_median['JobSatPoints_7'], bar_width, bottom=job_satisfaction_median['JobSatPoints_6'], label='JobSatPoints_7', color='purple')

ax.set_xlabel('Employment Groups', fontsize=14)
ax.set_ylabel('Median JobSat Scores', fontsize=14)
ax.set_title('Median JobSat 6 & 7 for Employment Groups', fontsize=16, fontweight='bold')
ax.set_xticks(employment_groups)
ax.set_xticklabels(job_satisfaction_median.index, rotation=45, ha='right')
ax.legend()
plt.tight_layout()
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here

age_list = {
    'Under 18 years old': 17, '18-24 years old': 21, '25-34 years old': 32, '35-44 years old': 39,
    '45-54 years old': 49, '55-64 years old': 59, '65 years or older': 70, 'Prefer not to say': None
}
df['AgeNum'] = df['Age'].map(age_list)

age_filtered_df = df[(df['AgeNum'] >= 30) & (df['AgeNum'] <= 35)]
age_filtered_df = age_filtered_df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])
compensation_bins = [0, 50000, 100000, 150000, 200000, np.inf]
compensation_labels = ['< $50K', '$50K-$100K', '$100K-$150K', '$150K-$200K', '> $200K']
age_filtered_df['CompensationRange'] = pd.cut(age_filtered_df['ConvertedCompYearly'], bins=compensation_bins, labels=compensation_labels)

grouped_data = age_filtered_df.groupby('CompensationRange')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

fig, ax = plt.subplots(figsize=(8, 8))
bar_width = 0.5
compensation_ranges = np.arange(len(grouped_data))
bar1 = ax.bar(compensation_ranges, grouped_data['ConvertedCompYearly'], bar_width, label='Compensation', color='red')
bar2 = ax.bar(compensation_ranges, grouped_data['JobSatPoints_6'], bar_width, bottom=grouped_data['ConvertedCompYearly'], label='Job Satisfaction', color='purple')
ax.set_xlabel('Comp. Range', fontsize=14)
ax.set_ylabel('Median Values', fontsize=14)
ax.set_title('Comp. and JobSat Between Group Aged 30-35', fontsize=16, fontweight='bold')
ax.set_xticks(compensation_ranges)
ax.set_xticklabels(grouped_data.index)
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
age_mapping = {
    'Under 18 years old': 17, '18-24 years old': 21, '25-34 years old': 29, '35-44 years old': 39,
    '45-54 years old': 49, '55-64 years old': 59, '65 years or older': 70, 'Prefer not to say': None
}
age_labels = {
    17: 'Under 18 years old', 21: '18-24 years old', 29: '25-34 years old', 39: '35-44 years old',
    49: '45-54 years old', 59: '55-64 years old', 70: '65 years or older'
}
df['AgeNum'] = df['Age'].map(age_mapping)

df_filtered = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'AgeNum'])
compensation_bins = [0, 50000, 100000, 150000, 200000, np.inf]
compensation_labels = ['< $50K', '$50K-$100K', '$100K-$150K', '$150K-$200K', '> $200K']
df_filtered['CompensationRange'] = pd.cut(df_filtered['ConvertedCompYearly'], bins=compensation_bins, labels=compensation_labels)

grouped_data = df_filtered.groupby(['AgeNum', 'CompensationRange'])[['ConvertedCompYearly', 'JobSatPoints_6']].median().unstack()

fig, ax = plt.subplots(figsize=(8, 8))
bar_width = 0.35
age_groups = np.arange(len(grouped_data))
bottom = np.zeros(len(grouped_data))
colors = ['red', 'purple', 'yellow', 'green', 'blue']
for idx, comp_range in enumerate(compensation_labels):
    comp_values = grouped_data['ConvertedCompYearly'][comp_range]
    job_sat_values = grouped_data['JobSatPoints_6'][comp_range]
    ax.bar(age_groups, comp_values, bar_width, label=f'Compensation {comp_range}', color=colors[idx], bottom=bottom)
    bottom += comp_values

ax.set_xlabel('AgeGroups', fontsize=14)
ax.set_ylabel('Median Values', fontsize=14)
ax.set_title('Median Comp. and JobSat by All Age Groups', fontsize=16, fontweight='bold')
ax.set_xticks(age_groups)
ax.set_xticklabels([age_labels[age] for age in grouped_data.index])
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
##Write your code here

database_preferences = df['DatabaseWantToWorkWith'].str.split(';').explode()
db_by_age = database_preferences.groupby([df['Age_num'], database_preferences]).size().unstack(fill_value=0).iloc[:, :5]

db_by_age.plot(kind='bar', stacked=True, figsize=(8, 8), colormap='Spectral')
plt.title('DatabaseWantToWorkWith by All Age Groups')
plt.xlabel('AgeGroups')
plt.ylabel('No. of Respondants')
plt.legend(title='Databases', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
##Write your code here

top_employment_types = df_filtered['Employment'].value_counts().head(5).index
df_filtered = df_filtered[df_filtered['Employment'].isin(top_employment_types)]
grouped_data = df_filtered.groupby(['JobSat', 'Employment']).size().unstack().fillna(0)
grouped_data = grouped_data.div(grouped_data.sum(axis=1), axis=0)

fig, ax = plt.subplots(figsize=(14, 8))
bar_width = 0.85
job_satisfaction_levels = np.arange(len(grouped_data))
bottom = np.zeros(len(grouped_data))
colors = ['red', 'purple', 'yellow', 'green', 'blue']
for idx, employment_type in enumerate(top_employment_types):
    values = grouped_data[employment_type]
    bars = ax.bar(job_satisfaction_levels, values, bar_width, label=employment_type, color=colors[idx], bottom=bottom)
    bottom += values

    for bar_idx, bar in enumerate(bars):
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width() / 2., bottom[bar_idx] - height / 2.,
                    f'{height:.2f}', ha='center', va='center', fontsize=10, color='black')
ax.set_xlabel('JobSat Level', fontsize=14)
ax.set_ylabel('Proportions', fontsize=14)
ax.set_title('Employment Types by JobSat Level - Distribution', fontsize=16, fontweight='bold')
ax.set_xticks(job_satisfaction_levels)
ax.set_xticklabels(grouped_data.index)
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)
